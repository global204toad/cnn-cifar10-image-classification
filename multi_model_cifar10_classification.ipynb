```python
# ============================================================
# CIFAR-10 Multi-Model + Multimodal Image Classification
# ANN, CNNs, Transfer Learning, and Multimodal Models
# ============================================================

import os
import tarfile
import subprocess
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

print("TensorFlow Version:", tf.__version__)
print("GPU Available:", tf.config.list_physical_devices("GPU"))

keras.utils.set_random_seed(42)

# ============================================================
# 1. DOWNLOAD CIFAR-10 DATASET
# ============================================================

dataset_url = "https://s3.amazonaws.com/fast-ai-imageclas/cifar10.tgz"
dataset_file = "/content/cifar10.tgz" if os.path.exists("/content") else "cifar10.tgz"
extract_path = "/content/cifar10_data" if os.path.exists("/content") else "cifar10_data"

if not os.path.exists(dataset_file):
    print("Downloading CIFAR-10 dataset...")
    subprocess.run(["wget", "-O", dataset_file, dataset_url], check=True)
else:
    print("Dataset file already exists.")

if not os.path.exists(extract_path):
    print("Extracting dataset...")
    os.makedirs(extract_path, exist_ok=True)

    with tarfile.open(dataset_file, "r:gz") as tar:
        try:
            tar.extractall(path=extract_path, filter="data")
        except TypeError:
            tar.extractall(path=extract_path)
else:
    print("Dataset already extracted.")

# Find train and test folders automatically
train_dir = None
test_dir = None

for root, dirs, files in os.walk(extract_path):
    if "train" in dirs and "test" in dirs:
        train_dir = os.path.join(root, "train")
        test_dir = os.path.join(root, "test")
        break

if train_dir is None or test_dir is None:
    raise Exception("Could not find train/test folders. Dataset extraction problem.")

print("Train directory:", train_dir)
print("Test directory:", test_dir)

# ============================================================
# 2. LOAD DATASET
# ============================================================

IMG_SIZE = (32, 32)
BATCH_SIZE = 128
AUTOTUNE = tf.data.AUTOTUNE

train_ds_raw = keras.utils.image_dataset_from_directory(
    train_dir,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode="int",
    shuffle=True,
    seed=42
)

test_ds_raw = keras.utils.image_dataset_from_directory(
    test_dir,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode="int",
    shuffle=False
)

class_names = train_ds_raw.class_names
print("Class names:", class_names)

train_ds = train_ds_raw.cache().prefetch(AUTOTUNE)
test_ds = test_ds_raw.cache().prefetch(AUTOTUNE)

# ============================================================
# 3. CREATE MULTIMODAL DATASET
# ============================================================
# CIFAR-10 is originally image-only.
# To make a correct multimodal model, we add a second input:
# numerical metadata extracted from the image.
#
# Input 1: image
# Input 2: metadata features
#
# Metadata features:
# - mean RGB
# - standard deviation RGB
# - brightness
# - contrast

def extract_image_metadata(images):
    images = tf.cast(images, tf.float32) / 255.0

    mean_rgb = tf.reduce_mean(images, axis=[1, 2])
    std_rgb = tf.math.reduce_std(images, axis=[1, 2])

    grayscale = tf.reduce_mean(images, axis=-1, keepdims=True)
    brightness = tf.reduce_mean(grayscale, axis=[1, 2, 3], keepdims=False)
    contrast = tf.math.reduce_std(grayscale, axis=[1, 2, 3], keepdims=False)

    brightness = tf.expand_dims(brightness, axis=-1)
    contrast = tf.expand_dims(contrast, axis=-1)

    metadata = tf.concat(
        [mean_rgb, std_rgb, brightness, contrast],
        axis=-1
    )

    return metadata


def add_metadata(images, labels):
    metadata = extract_image_metadata(images)

    inputs = {
        "image_input": images,
        "metadata_input": metadata
    }

    return inputs, labels


train_multi_ds = train_ds_raw.map(add_metadata).cache().prefetch(AUTOTUNE)
test_multi_ds = test_ds_raw.map(add_metadata).cache().prefetch(AUTOTUNE)

# ============================================================
# 4. SHOW SAMPLE IMAGES
# ============================================================

plt.figure(figsize=(10, 5))

for images, labels in train_ds.take(1):
    for i in range(10):
        plt.subplot(2, 5, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i].numpy()])
        plt.axis("off")

plt.suptitle("Sample Images from CIFAR-10 Dataset")
plt.tight_layout()
plt.show()

# ============================================================
# 5. HELPER FUNCTIONS
# ============================================================

def create_data_augmentation():
    return keras.Sequential([
        layers.RandomFlip("horizontal"),
        layers.RandomRotation(0.08),
        layers.RandomZoom(0.10),
        layers.RandomTranslation(0.08, 0.08)
    ], name="data_augmentation")


def compile_model(model, learning_rate=0.001):
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=learning_rate),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model


def plot_history(history, model_name):
    plt.figure(figsize=(8, 5))
    plt.plot(history.history["accuracy"], label="Training Accuracy")
    plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title(model_name + " Accuracy")
    plt.legend()
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(8, 5))
    plt.plot(history.history["loss"], label="Training Loss")
    plt.plot(history.history["val_loss"], label="Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(model_name + " Loss")
    plt.legend()
    plt.grid(True)
    plt.show()


def evaluate_model(model, model_name, test_data):
    loss, accuracy = model.evaluate(test_data, verbose=0)
    print(f"{model_name} Test Loss: {loss:.4f}")
    print(f"{model_name} Test Accuracy: {accuracy:.4f}")
    return loss, accuracy


def train_model(model, model_name, train_data, test_data, epochs, learning_rate=0.001, patience=4):
    print("\n==============================")
    print(f"Training: {model_name}")
    print("==============================")

    compile_model(model, learning_rate=learning_rate)
    model.summary()

    callbacks = [
        keras.callbacks.EarlyStopping(
            monitor="val_accuracy",
            patience=patience,
            restore_best_weights=True
        )
    ]

    history = model.fit(
        train_data,
        epochs=epochs,
        validation_data=test_data,
        callbacks=callbacks,
        verbose=1
    )

    loss, accuracy = evaluate_model(model, model_name, test_data)
    plot_history(history, model_name)

    return history, loss, accuracy


def get_true_and_predicted_labels(model, test_data):
    y_true = []
    y_pred = []

    for batch_inputs, labels in test_data:
        predictions = model.predict(batch_inputs, verbose=0)
        predicted_labels = np.argmax(predictions, axis=1)

        y_true.extend(labels.numpy())
        y_pred.extend(predicted_labels)

    return np.array(y_true), np.array(y_pred)


def show_predictions(model, model_name, test_data):
    for batch_inputs, labels in test_data.take(1):
        predictions = model.predict(batch_inputs, verbose=0)
        predicted_labels = np.argmax(predictions, axis=1)

        if isinstance(batch_inputs, dict):
            images = batch_inputs["image_input"]
        else:
            images = batch_inputs

        plt.figure(figsize=(12, 8))

        for i in range(12):
            plt.subplot(3, 4, i + 1)
            plt.imshow(images[i].numpy().astype("uint8"))

            true_label = class_names[labels[i].numpy()]
            predicted_label = class_names[predicted_labels[i]]

            color = "green" if true_label == predicted_label else "red"

            plt.title(
                f"True: {true_label}\nPred: {predicted_label}",
                color=color
            )
            plt.axis("off")

        plt.suptitle(model_name + " Prediction Results")
        plt.tight_layout()
        plt.show()


# ============================================================
# MODEL 1: ANN
# ============================================================

ann_model = keras.Sequential([
    layers.Input(shape=(32, 32, 3)),
    layers.Rescaling(1.0 / 255),
    layers.Flatten(),

    layers.Dense(512, activation="relu"),
    layers.Dropout(0.3),

    layers.Dense(256, activation="relu"),
    layers.Dropout(0.3),

    layers.Dense(10, activation="softmax")
], name="ANN")

ann_history, ann_loss, ann_accuracy = train_model(
    ann_model,
    "ANN",
    train_ds,
    test_ds,
    epochs=8,
    learning_rate=0.001
)

# ============================================================
# MODEL 2: SIMPLE CNN
# ============================================================

simple_cnn_model = keras.Sequential([
    layers.Input(shape=(32, 32, 3)),
    layers.Rescaling(1.0 / 255),

    layers.Conv2D(32, (3, 3), activation="relu"),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation="relu"),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),

    layers.Dense(128, activation="relu"),
    layers.Dropout(0.3),

    layers.Dense(10, activation="softmax")
], name="Simple_CNN")

simple_cnn_history, simple_cnn_loss, simple_cnn_accuracy = train_model(
    simple_cnn_model,
    "Simple CNN",
    train_ds,
    test_ds,
    epochs=10,
    learning_rate=0.001
)

# ============================================================
# MODEL 3: IMPROVED CNN
# ============================================================

improved_cnn_model = keras.Sequential([
    layers.Input(shape=(32, 32, 3)),
    layers.Rescaling(1.0 / 255),

    layers.Conv2D(32, (3, 3), activation="relu", padding="same"),
    layers.BatchNormalization(),
    layers.Conv2D(32, (3, 3), activation="relu", padding="same"),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(64, (3, 3), activation="relu", padding="same"),
    layers.BatchNormalization(),
    layers.Conv2D(64, (3, 3), activation="relu", padding="same"),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(128, (3, 3), activation="relu", padding="same"),
    layers.BatchNormalization(),
    layers.Conv2D(128, (3, 3), activation="relu", padding="same"),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.35),

    layers.Flatten(),

    layers.Dense(256, activation="relu"),
    layers.BatchNormalization(),
    layers.Dropout(0.5),

    layers.Dense(10, activation="softmax")
], name="Improved_CNN")

improved_cnn_history, improved_cnn_loss, improved_cnn_accuracy = train_model(
    improved_cnn_model,
    "Improved CNN",
    train_ds,
    test_ds,
    epochs=15,
    learning_rate=0.001
)

# ============================================================
# MODEL 4: DEEP CNN
# ============================================================

deep_cnn_model = keras.Sequential([
    layers.Input(shape=(32, 32, 3)),
    layers.Rescaling(1.0 / 255),

    layers.Conv2D(32, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.Conv2D(32, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(64, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.Conv2D(64, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.30),

    layers.Conv2D(128, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.Conv2D(128, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.35),

    layers.Conv2D(256, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),

    layers.GlobalAveragePooling2D(),

    layers.Dense(256, activation="relu"),
    layers.BatchNormalization(),
    layers.Dropout(0.5),

    layers.Dense(10, activation="softmax")
], name="Deep_CNN")

deep_cnn_history, deep_cnn_loss, deep_cnn_accuracy = train_model(
    deep_cnn_model,
    "Deep CNN",
    train_ds,
    test_ds,
    epochs=20,
    learning_rate=0.001
)

# ============================================================
# MODEL 5: CNN WITH DATA AUGMENTATION
# ============================================================

augmented_cnn_model = keras.Sequential([
    layers.Input(shape=(32, 32, 3)),
    create_data_augmentation(),
    layers.Rescaling(1.0 / 255),

    layers.Conv2D(32, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.Conv2D(32, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(64, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.Conv2D(64, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.30),

    layers.Conv2D(128, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.Conv2D(128, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.40),

    layers.GlobalAveragePooling2D(),

    layers.Dense(256, activation="relu"),
    layers.BatchNormalization(),
    layers.Dropout(0.5),

    layers.Dense(10, activation="softmax")
], name="CNN_With_Data_Augmentation")

augmented_cnn_history, augmented_cnn_loss, augmented_cnn_accuracy = train_model(
    augmented_cnn_model,
    "CNN with Data Augmentation",
    train_ds,
    test_ds,
    epochs=25,
    learning_rate=0.001
)

# ============================================================
# MODEL 6: TRANSFER LEARNING MobileNetV2
# ============================================================

mobilenet_base = MobileNetV2(
    input_shape=(96, 96, 3),
    include_top=False,
    weights="imagenet"
)

mobilenet_base.trainable = False

transfer_learning_model = keras.Sequential([
    layers.Input(shape=(32, 32, 3)),
    create_data_augmentation(),
    layers.Resizing(96, 96),
    layers.Lambda(preprocess_input),

    mobilenet_base,

    layers.GlobalAveragePooling2D(),
    layers.Dropout(0.4),

    layers.Dense(128, activation="relu"),
    layers.Dropout(0.3),

    layers.Dense(10, activation="softmax")
], name="Transfer_Learning_MobileNetV2")

transfer_history, transfer_loss, transfer_accuracy = train_model(
    transfer_learning_model,
    "Transfer Learning MobileNetV2",
    train_ds,
    test_ds,
    epochs=15,
    learning_rate=0.0005
)

# Fine-tuning MobileNetV2
print("\n==============================")
print("Fine-tuning MobileNetV2")
print("==============================")

mobilenet_base.trainable = True

for layer in mobilenet_base.layers[:-30]:
    layer.trainable = False

compile_model(transfer_learning_model, learning_rate=0.00005)

fine_tune_history = transfer_learning_model.fit(
    train_ds,
    epochs=8,
    validation_data=test_ds,
    callbacks=[
        keras.callbacks.EarlyStopping(
            monitor="val_accuracy",
            patience=3,
            restore_best_weights=True
        )
    ],
    verbose=1
)

transfer_loss, transfer_accuracy = evaluate_model(
    transfer_learning_model,
    "Transfer Learning MobileNetV2 Fine-Tuned",
    test_ds
)

plot_history(fine_tune_history, "Transfer Learning MobileNetV2 Fine-Tuned")

# ============================================================
# MODEL 7: MULTIMODAL CNN
# ============================================================
# This model uses two inputs:
# 1. Image input
# 2. Metadata input
#
# Image branch learns visual features.
# Metadata branch learns numerical features.
# Then both branches are combined.

image_input = keras.Input(shape=(32, 32, 3), name="image_input")
metadata_input = keras.Input(shape=(8,), name="metadata_input")

# Image branch
x = layers.Rescaling(1.0 / 255)(image_input)

x = layers.Conv2D(32, (3, 3), padding="same", activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.Conv2D(32, (3, 3), padding="same", activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.MaxPooling2D((2, 2))(x)
x = layers.Dropout(0.25)(x)

x = layers.Conv2D(64, (3, 3), padding="same", activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.Conv2D(64, (3, 3), padding="same", activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.MaxPooling2D((2, 2))(x)
x = layers.Dropout(0.30)(x)

x = layers.Conv2D(128, (3, 3), padding="same", activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.Conv2D(128, (3, 3), padding="same", activation="relu")(x)
x = layers.BatchNormalization()(x)
x = layers.MaxPooling2D((2, 2))(x)
x = layers.Dropout(0.35)(x)

x = layers.GlobalAveragePooling2D()(x)

# Metadata branch
m = layers.Dense(32, activation="relu")(metadata_input)
m = layers.BatchNormalization()(m)
m = layers.Dropout(0.2)(m)

m = layers.Dense(32, activation="relu")(m)

# Combine image features and metadata features
combined = layers.Concatenate()([x, m])

combined = layers.Dense(256, activation="relu")(combined)
combined = layers.BatchNormalization()(combined)
combined = layers.Dropout(0.5)(combined)

combined = layers.Dense(128, activation="relu")(combined)
combined = layers.Dropout(0.3)(combined)

output = layers.Dense(10, activation="softmax")(combined)

multimodal_cnn_model = keras.Model(
    inputs={
        "image_input": image_input,
        "metadata_input": metadata_input
    },
    outputs=output,
    name="Multimodal_CNN"
)

multimodal_cnn_history, multimodal_cnn_loss, multimodal_cnn_accuracy = train_model(
    multimodal_cnn_model,
    "Multimodal CNN",
    train_multi_ds,
    test_multi_ds,
    epochs=25,
    learning_rate=0.001
)

# ============================================================
# MODEL 8: MULTIMODAL MobileNetV2
# ============================================================
# This is the strongest multimodal model.
# It combines:
# - pretrained MobileNetV2 image features
# - numerical metadata features

multi_mobilenet_base = MobileNetV2(
    input_shape=(96, 96, 3),
    include_top=False,
    weights="imagenet"
)

multi_mobilenet_base.trainable = False

image_input = keras.Input(shape=(32, 32, 3), name="image_input")
metadata_input = keras.Input(shape=(8,), name="metadata_input")

# Image branch using MobileNetV2
x = create_data_augmentation()(image_input)
x = layers.Resizing(96, 96)(x)
x = layers.Lambda(preprocess_input)(x)
x = multi_mobilenet_base(x)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.4)(x)

# Metadata branch
m = layers.Dense(32, activation="relu")(metadata_input)
m = layers.BatchNormalization()(m)
m = layers.Dropout(0.2)(m)

m = layers.Dense(32, activation="relu")(m)

# Combine both modalities
combined = layers.Concatenate()([x, m])

combined = layers.Dense(256, activation="relu")(combined)
combined = layers.BatchNormalization()(combined)
combined = layers.Dropout(0.5)(combined)

combined = layers.Dense(128, activation="relu")(combined)
combined = layers.Dropout(0.3)(combined)

output = layers.Dense(10, activation="softmax")(combined)

multimodal_mobilenet_model = keras.Model(
    inputs={
        "image_input": image_input,
        "metadata_input": metadata_input
    },
    outputs=output,
    name="Multimodal_MobileNetV2"
)

multimodal_mobilenet_history, multimodal_mobilenet_loss, multimodal_mobilenet_accuracy = train_model(
    multimodal_mobilenet_model,
    "Multimodal MobileNetV2",
    train_multi_ds,
    test_multi_ds,
    epochs=15,
    learning_rate=0.0005
)

# Fine-tuning Multimodal MobileNetV2
print("\n==============================")
print("Fine-tuning Multimodal MobileNetV2")
print("==============================")

multi_mobilenet_base.trainable = True

for layer in multi_mobilenet_base.layers[:-30]:
    layer.trainable = False

compile_model(multimodal_mobilenet_model, learning_rate=0.00005)

multimodal_fine_tune_history = multimodal_mobilenet_model.fit(
    train_multi_ds,
    epochs=8,
    validation_data=test_multi_ds,
    callbacks=[
        keras.callbacks.EarlyStopping(
            monitor="val_accuracy",
            patience=3,
            restore_best_weights=True
        )
    ],
    verbose=1
)

multimodal_mobilenet_loss, multimodal_mobilenet_accuracy = evaluate_model(
    multimodal_mobilenet_model,
    "Multimodal MobileNetV2 Fine-Tuned",
    test_multi_ds
)

plot_history(multimodal_fine_tune_history, "Multimodal MobileNetV2 Fine-Tuned")

# ============================================================
# MODEL COMPARISON
# ============================================================

results = pd.DataFrame({
    "Model": [
        "ANN",
        "Simple CNN",
        "Improved CNN",
        "Deep CNN",
        "CNN with Data Augmentation",
        "Transfer Learning MobileNetV2",
        "Multimodal CNN",
        "Multimodal MobileNetV2"
    ],
    "Test Loss": [
        ann_loss,
        simple_cnn_loss,
        improved_cnn_loss,
        deep_cnn_loss,
        augmented_cnn_loss,
        transfer_loss,
        multimodal_cnn_loss,
        multimodal_mobilenet_loss
    ],
    "Test Accuracy": [
        ann_accuracy,
        simple_cnn_accuracy,
        improved_cnn_accuracy,
        deep_cnn_accuracy,
        augmented_cnn_accuracy,
        transfer_accuracy,
        multimodal_cnn_accuracy,
        multimodal_mobilenet_accuracy
    ]
})

print("\n==============================")
print("Model Comparison Results")
print("==============================")
print(results)

plt.figure(figsize=(14, 6))
plt.bar(results["Model"], results["Test Accuracy"])
plt.xlabel("Model")
plt.ylabel("Test Accuracy")
plt.title("CIFAR-10 Model Accuracy Comparison")
plt.ylim(0, 1)
plt.xticks(rotation=35, ha="right")
plt.grid(axis="y")
plt.tight_layout()
plt.show()

# ============================================================
# BEST MODEL EVALUATION
# ============================================================

models = {
    "ANN": (ann_model, test_ds),
    "Simple CNN": (simple_cnn_model, test_ds),
    "Improved CNN": (improved_cnn_model, test_ds),
    "Deep CNN": (deep_cnn_model, test_ds),
    "CNN with Data Augmentation": (augmented_cnn_model, test_ds),
    "Transfer Learning MobileNetV2": (transfer_learning_model, test_ds),
    "Multimodal CNN": (multimodal_cnn_model, test_multi_ds),
    "Multimodal MobileNetV2": (multimodal_mobilenet_model, test_multi_ds)
}

best_model_name = results.loc[results["Test Accuracy"].idxmax(), "Model"]
best_model, best_test_data = models[best_model_name]

print("\nBest Model:", best_model_name)

show_predictions(best_model, best_model_name, best_test_data)

y_true, y_pred = get_true_and_predicted_labels(best_model, best_test_data)

cm = confusion_matrix(y_true, y_pred)

fig, ax = plt.subplots(figsize=(10, 10))
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=class_names
)

disp.plot(cmap="Blues", xticks_rotation=45, ax=ax)
plt.title("Confusion Matrix for Best Model: " + best_model_name)
plt.tight_layout()
plt.show()

print("\n==============================")
print("Classification Report for Best Model")
print("==============================")

print(
    classification_report(
        y_true,
        y_pred,
        target_names=class_names
    )
)

# ============================================================
# SAVE MODELS AND RESULTS
# ============================================================

ann_model.save("ann_cifar10_model.keras")
simple_cnn_model.save("simple_cnn_cifar10_model.keras")
improved_cnn_model.save("improved_cnn_cifar10_model.keras")
deep_cnn_model.save("deep_cnn_cifar10_model.keras")
augmented_cnn_model.save("augmented_cnn_cifar10_model.keras")
transfer_learning_model.save("mobilenetv2_transfer_cifar10_model.keras")
multimodal_cnn_model.save("multimodal_cnn_cifar10_model.keras")
multimodal_mobilenet_model.save("multimodal_mobilenetv2_cifar10_model.keras")

results.to_csv("cifar10_model_comparison_results.csv", index=False)

print("\nAll models saved successfully.")
print("Results saved to cifar10_model_comparison_results.csv")
print("Project completed successfully.")
```
