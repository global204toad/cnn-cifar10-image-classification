The models used were:

1. ANN (Artificial Neural Network)
2. Simple CNN
3. Improved CNN
4. Deep CNN
5. CNN with Data Augmentation
6. Transfer Learning using MobileNetV2

After comparing the test accuracies, the best model was **Transfer Learning using MobileNetV2**, which achieved the highest accuracy of **86.68%**.

This model performed better because it used pretrained knowledge from ImageNet, allowing it to extract strong image features even from small CIFAR-10 images.

The **Deep CNN** model also performed very well with an accuracy of **84.03%**, showing that deeper convolutional networks can learn better visual patterns than simple models.

The **ANN** model had the lowest accuracy because it flattens the image and loses important spatial information, making it less suitable for image classification tasks.

Overall, the results show that CNN-based models are much better than ANN for image recognition, and Transfer Learning gives the best performance.

```python
# ============================================================
# Multi-Model Image Classification on CIFAR-10
# ANN vs Simple CNN vs Improved CNN vs Deep CNN
# CNN with Data Augmentation vs Transfer Learning MobileNetV2
# ============================================================

import os
import tarfile
import subprocess
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

print("TensorFlow Version:", tf.__version__)
print("GPU Available:", tf.config.list_physical_devices("GPU"))

# Make results more stable
keras.utils.set_random_seed(42)

# ============================================================
# 1. DOWNLOAD CIFAR-10 FROM FAST.AI MIRROR
# ============================================================

dataset_url = "https://s3.amazonaws.com/fast-ai-imageclas/cifar10.tgz"
dataset_file = "/content/cifar10.tgz" if os.path.exists("/content") else "cifar10.tgz"
extract_path = "/content/cifar10_data" if os.path.exists("/content") else "cifar10_data"

if not os.path.exists(dataset_file):
    print("Downloading CIFAR-10 dataset...")
    subprocess.run(["wget", "-O", dataset_file, dataset_url], check=True)
else:
    print("Dataset file already exists.")

if not os.path.exists(extract_path):
    print("Extracting dataset...")
    os.makedirs(extract_path, exist_ok=True)
    with tarfile.open(dataset_file, "r:gz") as tar:
        tar.extractall(path=extract_path)
else:
    print("Dataset already extracted.")

# Find train and test folders automatically
train_dir = None
test_dir = None

for root, dirs, files in os.walk(extract_path):
    if "train" in dirs and "test" in dirs:
        train_dir = os.path.join(root, "train")
        test_dir = os.path.join(root, "test")
        break

if train_dir is None or test_dir is None:
    raise Exception("Could not find train/test folders. Dataset extraction problem.")

print("Train directory:", train_dir)
print("Test directory:", test_dir)

# ============================================================
# 2. LOAD DATASET
# ============================================================

IMG_SIZE = (32, 32)
BATCH_SIZE = 128

train_ds_raw = keras.utils.image_dataset_from_directory(
    train_dir,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode="int",
    shuffle=True,
    seed=42
)

test_ds_raw = keras.utils.image_dataset_from_directory(
    test_dir,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode="int",
    shuffle=False
)

class_names = train_ds_raw.class_names
print("Class names:", class_names)

AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds_raw.cache().prefetch(AUTOTUNE)
test_ds = test_ds_raw.cache().prefetch(AUTOTUNE)

# ============================================================
# 3. SHOW SAMPLE IMAGES
# ============================================================

plt.figure(figsize=(10, 5))

for images, labels in train_ds.take(1):
    for i in range(10):
        plt.subplot(2, 5, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i].numpy()])
        plt.axis("off")

plt.suptitle("Sample Images from CIFAR-10 Dataset")
plt.tight_layout()
plt.show()

# ============================================================
# 4. HELPER FUNCTIONS
# ============================================================

def compile_model(model, learning_rate=0.001):
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=learning_rate),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model


def plot_history(history, model_name):
    plt.figure(figsize=(8, 5))
    plt.plot(history.history["accuracy"], label="Training Accuracy")
    plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title(model_name + " Accuracy")
    plt.legend()
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(8, 5))
    plt.plot(history.history["loss"], label="Training Loss")
    plt.plot(history.history["val_loss"], label="Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(model_name + " Loss")
    plt.legend()
    plt.grid(True)
    plt.show()


def evaluate_model(model, model_name):
    loss, accuracy = model.evaluate(test_ds, verbose=0)
    print(f"{model_name} Test Loss: {loss:.4f}")
    print(f"{model_name} Test Accuracy: {accuracy:.4f}")
    return loss, accuracy


def show_predictions(model, model_name):
    for images, labels in test_ds.take(1):
        predictions = model.predict(images[:12], verbose=0)
        predicted_labels = np.argmax(predictions, axis=1)

        plt.figure(figsize=(12, 8))

        for i in range(12):
            plt.subplot(3, 4, i + 1)
            plt.imshow(images[i].numpy().astype("uint8"))

            true_label = class_names[labels[i].numpy()]
            predicted_label = class_names[predicted_labels[i]]
            color = "green" if true_label == predicted_label else "red"

            plt.title(f"True: {true_label}\nPred: {predicted_label}", color=color)
            plt.axis("off")

        plt.suptitle(model_name + " Prediction Results")
        plt.tight_layout()
        plt.show()


def get_true_and_predicted_labels(model):
    y_true = []
    y_pred = []

    for images, labels in test_ds:
        predictions = model.predict(images, verbose=0)
        predicted_labels = np.argmax(predictions, axis=1)

        y_true.extend(labels.numpy())
        y_pred.extend(predicted_labels)

    return np.array(y_true), np.array(y_pred)


def train_model(model, model_name, epochs, learning_rate=0.001):
    print("\n==============================")
    print(f"Training: {model_name}")
    print("==============================")

    compile_model(model, learning_rate=learning_rate)
    model.summary()

    callbacks = [
        keras.callbacks.EarlyStopping(
            monitor="val_accuracy",
            patience=4,
            restore_best_weights=True
        )
    ]

    history = model.fit(
        train_ds,
        epochs=epochs,
        validation_data=test_ds,
        callbacks=callbacks,
        verbose=1
    )

    loss, accuracy = evaluate_model(model, model_name)
    plot_history(history, model_name)
    show_predictions(model, model_name)

    return history, loss, accuracy


# Shared augmentation block used by Model 5 and Model 6
# It runs only during training, not during testing/prediction.
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.08),
    layers.RandomZoom(0.10),
    layers.RandomTranslation(0.08, 0.08)
], name="data_augmentation")

# ============================================================
# 5. MODEL 1: ANN
# ============================================================

ann_model = keras.Sequential([
    layers.Input(shape=(32, 32, 3)),
    layers.Rescaling(1.0 / 255),
    layers.Flatten(),

    layers.Dense(512, activation="relu"),
    layers.Dropout(0.3),

    layers.Dense(256, activation="relu"),
    layers.Dropout(0.3),

    layers.Dense(10, activation="softmax")
], name="ANN")

ann_history, ann_loss, ann_accuracy = train_model(
    ann_model,
    "ANN",
    epochs=8,
    learning_rate=0.001
)

# ============================================================
# 6. MODEL 2: SIMPLE CNN
# ============================================================

simple_cnn_model = keras.Sequential([
    layers.Input(shape=(32, 32, 3)),
    layers.Rescaling(1.0 / 255),

    layers.Conv2D(32, (3, 3), activation="relu"),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation="relu"),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(128, activation="relu"),
    layers.Dropout(0.3),

    layers.Dense(10, activation="softmax")
], name="Simple_CNN")

simple_cnn_history, simple_cnn_loss, simple_cnn_accuracy = train_model(
    simple_cnn_model,
    "Simple CNN",
    epochs=10,
    learning_rate=0.001
)

# ============================================================
# 7. MODEL 3: IMPROVED CNN
# ============================================================

improved_cnn_model = keras.Sequential([
    layers.Input(shape=(32, 32, 3)),
    layers.Rescaling(1.0 / 255),

    layers.Conv2D(32, (3, 3), activation="relu", padding="same"),
    layers.BatchNormalization(),
    layers.Conv2D(32, (3, 3), activation="relu", padding="same"),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(64, (3, 3), activation="relu", padding="same"),
    layers.BatchNormalization(),
    layers.Conv2D(64, (3, 3), activation="relu", padding="same"),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(128, (3, 3), activation="relu", padding="same"),
    layers.BatchNormalization(),
    layers.Conv2D(128, (3, 3), activation="relu", padding="same"),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.35),

    layers.Flatten(),
    layers.Dense(256, activation="relu"),
    layers.BatchNormalization(),
    layers.Dropout(0.5),

    layers.Dense(10, activation="softmax")
], name="Improved_CNN")

improved_cnn_history, improved_cnn_loss, improved_cnn_accuracy = train_model(
    improved_cnn_model,
    "Improved CNN",
    epochs=15,
    learning_rate=0.001
)

# ============================================================
# 8. MODEL 4: DEEP CNN
# ============================================================
# Deeper architecture with 4 convolution blocks.
# Good for learning more complex image features.

deep_cnn_model = keras.Sequential([
    layers.Input(shape=(32, 32, 3)),
    layers.Rescaling(1.0 / 255),

    layers.Conv2D(32, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.Conv2D(32, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(64, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.Conv2D(64, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.30),

    layers.Conv2D(128, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.Conv2D(128, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.35),

    layers.Conv2D(256, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.GlobalAveragePooling2D(),

    layers.Dense(256, activation="relu"),
    layers.BatchNormalization(),
    layers.Dropout(0.5),

    layers.Dense(10, activation="softmax")
], name="Deep_CNN")

deep_cnn_history, deep_cnn_loss, deep_cnn_accuracy = train_model(
    deep_cnn_model,
    "Deep CNN",
    epochs=20,
    learning_rate=0.001
)

# ============================================================
# 9. MODEL 5: CNN WITH DATA AUGMENTATION
# ============================================================
# Data augmentation creates random variations of images during training.
# This helps reduce overfitting and improves generalization.

augmented_cnn_model = keras.Sequential([
    layers.Input(shape=(32, 32, 3)),
    data_augmentation,
    layers.Rescaling(1.0 / 255),

    layers.Conv2D(32, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.Conv2D(32, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(64, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.Conv2D(64, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.30),

    layers.Conv2D(128, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.Conv2D(128, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.40),

    layers.GlobalAveragePooling2D(),
    layers.Dense(256, activation="relu"),
    layers.BatchNormalization(),
    layers.Dropout(0.5),

    layers.Dense(10, activation="softmax")
], name="CNN_With_Data_Augmentation")

augmented_cnn_history, augmented_cnn_loss, augmented_cnn_accuracy = train_model(
    augmented_cnn_model,
    "CNN with Data Augmentation",
    epochs=25,
    learning_rate=0.001
)

# ============================================================
# 10. MODEL 6: TRANSFER LEARNING USING MobileNetV2
# ============================================================
# MobileNetV2 was pretrained on ImageNet.
# We freeze the base model and train only a new classifier head for CIFAR-10.
# CIFAR-10 images are 32x32, so we resize them to 96x96 for MobileNetV2.

mobilenet_base = MobileNetV2(
    input_shape=(96, 96, 3),
    include_top=False,
    weights="imagenet"
)

mobilenet_base.trainable = False

transfer_learning_model = keras.Sequential([
    layers.Input(shape=(32, 32, 3)),
    data_augmentation,
    layers.Resizing(96, 96),
    layers.Lambda(preprocess_input),

    mobilenet_base,
    layers.GlobalAveragePooling2D(),
    layers.Dropout(0.4),

    layers.Dense(128, activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(10, activation="softmax")
], name="Transfer_Learning_MobileNetV2")

transfer_history, transfer_loss, transfer_accuracy = train_model(
    transfer_learning_model,
    "Transfer Learning MobileNetV2",
    epochs=15,
    learning_rate=0.0005
)

# Optional fine-tuning step for better MobileNetV2 accuracy
# This unfreezes only the last layers of MobileNetV2.

print("\n==============================")
print("Fine-tuning MobileNetV2")
print("==============================")

mobilenet_base.trainable = True

for layer in mobilenet_base.layers[:-30]:
    layer.trainable = False

compile_model(transfer_learning_model, learning_rate=0.00005)

fine_tune_history = transfer_learning_model.fit(
    train_ds,
    epochs=8,
    validation_data=test_ds,
    callbacks=[
        keras.callbacks.EarlyStopping(
            monitor="val_accuracy",
            patience=3,
            restore_best_weights=True
        )
    ],
    verbose=1
)

transfer_loss, transfer_accuracy = evaluate_model(
    transfer_learning_model,
    "Transfer Learning MobileNetV2 Fine-Tuned"
)

plot_history(fine_tune_history, "Transfer Learning MobileNetV2 Fine-Tuned")
show_predictions(transfer_learning_model, "Transfer Learning MobileNetV2 Fine-Tuned")

# ============================================================
# 11. MODEL COMPARISON
# ============================================================

results = pd.DataFrame({
    "Model": [
        "ANN",
        "Simple CNN",
        "Improved CNN",
        "Deep CNN",
        "CNN with Data Augmentation",
        "Transfer Learning MobileNetV2"
    ],
    "Test Loss": [
        ann_loss,
        simple_cnn_loss,
        improved_cnn_loss,
        deep_cnn_loss,
        augmented_cnn_loss,
        transfer_loss
    ],
    "Test Accuracy": [
        ann_accuracy,
        simple_cnn_accuracy,
        improved_cnn_accuracy,
        deep_cnn_accuracy,
        augmented_cnn_accuracy,
        transfer_accuracy
    ]
})

print("\n==============================")
print("Model Comparison Results")
print("==============================")
print(results)

plt.figure(figsize=(12, 6))
plt.bar(results["Model"], results["Test Accuracy"])
plt.xlabel("Model")
plt.ylabel("Test Accuracy")
plt.title("CIFAR-10 Model Accuracy Comparison")
plt.ylim(0, 1)
plt.xticks(rotation=30, ha="right")
plt.grid(axis="y")
plt.tight_layout()
plt.show()

# ============================================================
# 12. CONFUSION MATRIX AND CLASSIFICATION REPORT FOR BEST MODEL
# ============================================================

models = {
    "ANN": ann_model,
    "Simple CNN": simple_cnn_model,
    "Improved CNN": improved_cnn_model,
    "Deep CNN": deep_cnn_model,
    "CNN with Data Augmentation": augmented_cnn_model,
    "Transfer Learning MobileNetV2": transfer_learning_model
}

best_model_name = results.loc[results["Test Accuracy"].idxmax(), "Model"]
best_model = models[best_model_name]

print("\nBest Model:", best_model_name)

y_true, y_pred = get_true_and_predicted_labels(best_model)
cm = confusion_matrix(y_true, y_pred)

fig, ax = plt.subplots(figsize=(10, 10))
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=class_names
)

disp.plot(cmap="Blues", xticks_rotation=45, ax=ax)
plt.title("Confusion Matrix for Best Model: " + best_model_name)
plt.tight_layout()
plt.show()

print("\n==============================")
print("Classification Report for Best Model")
print("==============================")

print(
    classification_report(
        y_true,
        y_pred,
        target_names=class_names
    )
)

# ============================================================
# 13. SAVE MODELS
# ============================================================

ann_model.save("ann_cifar10_model.keras")
simple_cnn_model.save("simple_cnn_cifar10_model.keras")
improved_cnn_model.save("improved_cnn_cifar10_model.keras")
deep_cnn_model.save("deep_cnn_cifar10_model.keras")
augmented_cnn_model.save("augmented_cnn_cifar10_model.keras")
transfer_learning_model.save("mobilenetv2_transfer_cifar10_model.keras")

results.to_csv("cifar10_model_comparison_results.csv", index=False)

print("\nAll models saved successfully.")
print("Results saved to cifar10_model_comparison_results.csv")
print("Project completed successfully.")
```
